```python
# Rglage du seed (optionnel)
np.random.seed(101)

# Colonnes d'actions (Stocks)
print('Stocks')
print(pfc_value.columns)
print('\n')

# Crer des poids alatoires
print('Creating Random Weights')
weights = np.array(np.random.random(4))
print(weights)
print('\n')

# Re-pondration des Poids
print('Rebalance to sum to 1.0')
weights = weights / np.sum(weights)
print(weights)
print('\n')

# Select the relevant columns from log_ret
relevant_log_retd = log_retd[['VOW3 Pos', 'PAH3 Pos', 'MBG Pos', 'CON Pos']]

# Rendement attendu ou escompt
print('Expected Portfolio Return')
exp_ret = np.sum(relevant_log_retd.mean()* weights) *252
print(exp_ret)
print('\n')

# variance attendue ou escompte
print('Expected Volatility')
exp_vol = np.sqrt(np.dot(weights.T, np.dot(relevant_log_retd.cov() * 252, weights)))
print(exp_vol)
print('\n')

# Ratio de Sharpe
SR = exp_ret/exp_vol
print('Sharpe Ratio')
print(SR)
```

```python
num_ports = 15000

# Only consider the 4 stock position columns
stock_columns = ['VOW3 Pos', 'PAH3 Pos', 'MBG Pos', 'CON Pos']
all_weights = np.zeros((num_ports, len(stock_columns)))  # Change here to 4 columns
retd_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for ind in range(num_ports):

    # Crer des poids alatoires
    weights = np.array(np.random.random(4))

    # (re)Pondration des poids
    weights = weights / np.sum(weights)

    # Sauvegarde des Poids
    all_weights[ind,:] = weights

    # rendement attendu
    retd_arr[ind] = np.sum((relevant_log_retd.mean() * weights) *252)

    # variance Attendue
    vol_arr[ind] = np.sqrt(np.dot(weights.T, np.dot(relevant_log_retd.cov() * 252, weights)))

    # Ratio de Sharpe
    sharpe_arr[ind] = retd_arr[ind]/vol_arr[ind]
```
