```python
for stock_df in (SAN, BN, SU, EN, RI, CA, DG, CAP, ML, VIV,
           ENGI, HO, VIE, ACA, SGO, SW, ORA, GLE, AI,
           BNP, LR, AIR, AC, WLN, ATO, OR, MC, KER):
    stock_df['Normed Return'] = stock_df['Close']/stock_df.iloc[0]['Close']
```

```python
for stock_df,allo in zip([SAN, BN, SU, EN, RI, CA, DG, CAP, ML, VIV,
           ENGI, HO, VIE, ACA, SGO, SW, ORA, GLE, AI,
           BNP, LR, AIR, AC, WLN, ATO, OR, MC, KER],[.3,.2,.1,.4]):
    stock_df['Allocation'] = stock_df['Normed Return']*allo
```

```python
for stock_df in [SAN, BN, SU, EN]:
    stock_df['Position Values'] = stock_df['Allocation']*1500
```

```python
all_pos_vals2 = [SAN['Position Values'],BN['Position Values'],SU['Position Values'],EN['Position Values']]
```

```python
portfolio_val2 = pd.concat(all_pos_vals2,axis=1)
```

```python
portfolio_val2.head()
```

```python
portfolio_val2.columns = ['SAN Pos','BN Pos','SU Pos','EN Pos']
```

```python
portfolio_val2.tail()
```

```python
portfolio_val2['Total Pos'] = portfolio_val2.sum(axis=1)
```

```python
portfolio_val2.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
portfolio_val2['Total Pos'].plot(figsize=(10,8))
plt.title('Valeur totale du portefeuille')
```

```python
portfolio_val2['Daily Return'] = portfolio_val2['Total Pos'].pct_change(1)
```
